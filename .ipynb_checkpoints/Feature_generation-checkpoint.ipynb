```python
import pandas as pd 
import geopandas
import tensorflow as tf 
import numpy as np 
from matplotlib import pyplot

from shapely.geometry import Point
from shapely.geometry import box
```

## Importing the Data 

```python
data = pd.read_csv("/Users/chamalgomes/Documents/Python/Kaggle_Comp/EY_Comp_Data/data_train.csv")
df = pd.DataFrame(data)
df.head()
```

## Feature 1: Time difference between entry and exit for one trajectory 

```python
df[['time_entry','time_exit']] = df[['time_entry','time_exit']].apply(pd.to_datetime)
df['time_diff'] = df['time_exit']-df['time_entry']
```

## Feature 2: Direction of Movement for each trajectory 

During one trajectory whether the vehicle moves towards or moves away from the city. 1 denotes moving towards and 0 denotes moving away.  

```python
gdf = geopandas.GeoDataFrame()
gdf1 = geopandas.GeoDataFrame()

b = box(3750901.5068, -19268905.6133, 3770901.5068, -19208905.6133)

gdf['Entry_Coordinates'] = list(zip(df.x_entry, df.y_entry))
gdf['Entry_Coordinates'] = gdf['Entry_Coordinates'].apply(Point)
gdf = geopandas.GeoDataFrame(gdf, geometry='Entry_Coordinates')

gdf1['Exit_Coordinates'] = list(zip(df.x_exit, df.y_exit))
gdf1['Exit_Coordinates'] = gdf1['Exit_Coordinates'].apply(Point)
gdf1 = geopandas.GeoDataFrame(gdf1, geometry='Exit_Coordinates')

dis_exit = gdf1['Exit_Coordinates'].distance(b)
dis_entry = gdf['Entry_Coordinates'].distance(b)

movement_direction = np.zeros(df.shape[0])

direction 
for i in range(df.shape[0]):
    if dis_entry[i] > dis_exit[i] :
        movement_direction[i] = 1
    else:
        movement_direction[i] = 0

df["movement_direction"]=pd.DataFrame(movement_direction)

print("Done")
```
